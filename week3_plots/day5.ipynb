```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Plot multiple subplots in one figure.


```python
fig, axes = plt.subplots(1,2, figsize=(10,5))

axes[0].hist(tips['day'])
axes[0].set_title('Day Count')
axes[0].set_xlabel('Days')

sns.violinplot(x='day', y='tip', data=tips,ax=axes[1], palette='Pastel1')
axes[1].set_title("Tip by Day")
```

```python
sns.barplot(tips['sex'])
sns.set_style('whitegrid')
sns.set_palette('pastel')
```

Create two subplots: one histogram and one scatterplot.


```python
fig, axes = plt.subplots(1,2, figsize=(12,4))

sns.histplot(tips['day'], ax=axes[0])
sns.scatterplot(x='total_bill', y='tip', data=tips, color='black')
```

Apply whitegrid style to a seaborn plot.


```python
sns.scatterplot(x='day', y='tip',data=tips, palette='whitegrid', color='black')
```

Plot sales trends for 3 products in the same chart with legends.


```python
import numpy as np

np.random.seed(42)  # For reproducible results

# Generate dates for 12 months
dates = pd.date_range('2024-01-01', periods=12, freq='ME')  # 'ME' for month end

# Generate sales data for 3 products with different patterns
product_a_sales = np.random.randint(100, 500, 12) + np.arange(12) * 10  # Trending upward
product_b_sales = np.random.randint(80, 400, 12) + np.sin(np.arange(12)) * 50  # Seasonal pattern
product_c_sales = np.random.randint(150, 350, 12) - np.arange(12) * 5  # Slight downward trend

sales_data = pd.DataFrame({
    'Date': dates,
    'Product_A': product_a_sales,
    'Product_B': product_b_sales,
    'Product_C': product_c_sales
})
sales_data
```

```python
plt.plot(sales_data['Date'], sales_data['Product_A'], label='Product A')
plt.plot(sales_data['Date'], sales_data['Product_B'], label='Product B')
plt.plot(sales_data['Date'], sales_data['Product_C'], label='Product C')
plt.legend()
```

Create a subplot with 4 plots (2Ã—2 layout) showing different chart types.


```python
fig, axes = plt.subplots(2,2, figsize=(10,10))

sns.barplot(x='total_bill', y='tip',data=tips, ax=axes[0,0])
axes[0,0].set_title('Bar Plot', fontweight='bold')

sns.violinplot(tips['tip'], ax=axes[0,1])
axes[0,1].set_title('Violin Plot')

sns.countplot(tips['day'], ax=axes[1,0])
axes[1,0].set_title('Count Plot')

sns.stripplot(x='total_bill', y='tip', data=tips, jitter=True, dodge=True, hue='sex')
plt.tight_layout()
```
