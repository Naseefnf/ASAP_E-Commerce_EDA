```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

## Matplotlib

1. Simple Line Chart
    - Plot y =. Label the axes and add a grid.

```python
x = [1,2,3,4]
y = [i*2 for i in x]

plt.plot(x,y)
plt.xlabel('X vales')
plt.ylabel('Y values')
```

2. Multiple Lines on One Plot
    - Plot two datasets on the same plot (for example, y1 = , y2 = ). Add a legend.

```python
x = [1,2,3,4]
y1 = [i*2 for i in x]
y2 = [i*3 for i in x]

plt.plot(x,y1, label='y1')
plt.plot(x,y2, label='y2')
plt.legend()
```

3. Bar Chart
    - Create a bar chart showing sales of products: ["A", "B", "C", "D"] and values. Add value labels above the bars.

```python
products = ['Pen', 'Book', 'Bag', 'Pencil']
revenue = [320, 510, 390, 410]

chart = plt.bar(products, revenue)
plt.bar_label(chart, labels=revenue)

```

4. Horizontal Bar Chart
    - Plot the above products and sales as a horizontal bar chart. Change the bar color.

```python
chart = plt.barh(products, revenue, color='black')
plt.bar_label(chart, labels=revenue)
```

5. Pie Chart
    - Make a pie chart of market shares: Apple40, Samsung30, Huawei15, Others15. Show percentages, explode "Apple."

```python
labels = ['Apple', 'Samsung', 'Realme', 'Others']
sizes = [40, 30, 15, 15]
plt.pie(sizes,
        labels=labels,
        autopct="%0.2f%%",
        explode=[0.1,0,0,0])
```

6. Histogram

8. Subplots: 2x2 Grid

## Seaborn

1. Countplot (Univariate)
    - Use the 'tips' dataset to plot a countplot of the day column. Add a descriptive title.

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(tips['day'])
plt.bar_label
```

2. Barplot (Bivariate)
    - Plot the average tip amount per day in the 'tips' dataset. Use a color palette and add value annotations.

```python
ax = sns.barplot(x='day', y='tip', data=tips, palette='mako', estimator='mean', ci=None)
for bar in ax.containers:
    ax.bar_label(bar, fmt='%.2f')
```

3. Boxplot 
    

```python
df_melted = tips.melt(value_vars=['total_bill','tip'], var_name='variable', value_name='amount')
df_melted
```

```python
sns.boxplot(x='variable', y='amount', data=df_melted)
```

Boxplpot + stripplot
- Overlay a stripplot on a boxplot for total_bill by day from the 'tips' dataset, using jitter on the stripplot.

```python
sns.boxplot(x='day',y='total_bill', data=tips, palette='mako')
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, dodge=True)
```

4. Swarm Plot
    - Visualize all tip values by sex in the 'tips' dataset using a swarmplot. Try adding a color for each category.

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex', palette='Set2')
```

5. Violin Plot
    - Make a violinplot of tip by smoker status. Add split by sex (hue).

```python
sns.violinplot(x='smoker', y='tip', data=tips, hue='sex')
```

6. Pairplot
    - Use the 'iris' dataset. Make a pairplot with hue as species.

```python
iris = sns.load_dataset('iris')
iris.head(10)
```

```python
sns.pairplot(data=iris, hue='species')
```

7. Heatmap
    - Use 'flights' dataset: create a pivot table of 'month' vs 'year' for 'passengers', then plot a heatmap.

```python
flight = sns.load_dataset('flights')
flight.head(10)
```

```python
pivoted = flight.pivot(columns=['year'], index='month', values='passengers')
pivoted
```

```python
matrix = pivoted.corr(numeric_only=True)
sns.heatmap(matrix, annot=True)
```

8. Clustermap
    - On the same 'flights' pivot table, plot a clustermap.

```python
sns.clustermap(matrix)
```

19. Regression Plot
    - Use sns.regplot to show the relationship between total_bill and tip in 'tips'. Add color and marker customizations.

```python
sns.regplot(x='total_bill', y='tip', data=tips, marker='*')
```
