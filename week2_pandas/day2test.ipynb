```python
import numpy as np
import pandas as pd
```

```python
data = {"Name":["Ali","Sara","John","Meera"], "Age":[25,28,23,30], "Salary":[50000,60000,55000,70000]}
df = pd.DataFrame(data)
df
```

Select the "Name" column from this DataFrame:

```python
df["Name"]
```

Select the first 3 rows using both slicing and iloc[].

```python
df[:3]
```

```python
df.iloc[0:3]
```

Filter all rows where "Age" is greater than 25.

```python
df[df["Age"] > 25]
```

Select rows 1–3 and columns "Name" and "Salary" using iloc[].

```python
df.iloc[0:3, [0,2]]
```

Filter rows where "Salary" is between 55,000 and 70,000.

```python
df[(df["Salary"] > 55000) & (df["Salary"] < 70000)]
```

Select rows where "Age" is less than 30 and "Salary" is greater than 55,000.

```python
df[(df["Age"] < 30) & (df["Salary"] > 55000)]
```

Using boolean indexing, select all rows where the "Name" starts with "S" or "M".

```python

data = {
    "Name": ["Sara", "Ali", "Meera", "John", "Sam", "Mohan"],
    "Age": [28, 25, 30, 23, 26, 29],
    "Salary": [60000, 50000, 70000, 55000, 58000, 62000]
}

df = pd.DataFrame(data)
df
```

```python
df[df["Name"].str.startswith(("S","M"))]
```
