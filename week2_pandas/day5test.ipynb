Create two DataFrames of employees (EmpID, Name, Salary) and stack them vertically using pd.concat().

```python
df1 = pd.DataFrame({
    "EmpID": [101,102,103,104],
    "Name": ["Ram","Sam","Man","Ham"],
    "Salary": [20000,23000,45000,10000],
})

df1
```

```python
df2 = pd.DataFrame({
    "EmpID": [106,107,108,109],
    "Name": ["Mar","Mas","Nam","Mah"],
    "Salary": [24000,37000,24500,70000],
})

df2
```

```python
pd.concat((df1,df2))
```

Merge an Employee DataFrame and a Department DataFrame on Dept_ID.


```python
employees = pd.DataFrame({
    "EmpID": [101, 102, 103, 104, 105, 106],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Dept_ID": [1, 2, 1, 3, 2, 3],
    "Salary": [60000, 55000, 70000, 50000, 65000, 72000]
})

# Department DataFrame
departments = pd.DataFrame({
    "Dept_ID": [1, 2, 3],
    "Dept_Name": ["HR", "Finance", "IT"]
})
```

```python
employees, departments
```

```python
pd.merge(employees, departments, on="Dept_ID")
```

Group the Employee DataFrame by Dept_ID (or Dept_Name) and calculate the average salary.

```python
employees.groupby("Dept_ID")["Salary"].mean()
```
