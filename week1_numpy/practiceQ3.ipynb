```python
import numpy as np
```

Create a 3D array (2×3×4) filled with 7s.


```python
print(np.full((2,3,4),7))
```

Generate an array of numbers from 50 to 100 with a step of 5.


```python
arr = np.arange(50,101,5)
arr
```

Create an identity matrix of size 4.


```python
print(np.identity(4, dtype=int))
```

Generate a linearly spaced array between 0 and 1 with 11 points.


```python
print(np.linspace(0,1,11))
```

Replace all odd numbers in an array (1–20) with -1.


```python
arr = np.arange(1,21)
arr[arr % 2 !=0] = -1
print(arr)
```

Reverse the rows of a 3×3 matrix.


```python
arr = np.arange(1,10).reshape(3,3)
print(arr[::-1, :])
```

Reverse the cols of a 3×3 matrix.


```python
print(arr[:, ::-1])
```

Swap the first and last rows of a 4×4 matrix.


```python
arr = np.arange(1,17).reshape(4,4)
print(arr)
```

```python
arr[[0,-1]] = arr[[-1,0]]
print(arr)
```

Create an array of shape (3×3×3) with values 0–26.


```python
arr = np.arange(0,27).reshape(3,3,3)
print(arr)
```

Find the index of all elements greater than 50 in a random array of size 15 (1–100).


```python
np.random.seed(0)

arr = np.random.randint(1,100,15)
arr
```

```python
ind = np.where(arr > 50)
print(ind)
```

Reshape numbers 1–24 into a 4×6 matrix and extract the 2nd row.


```python
arr1 = np.arange(1,25).reshape(4,6)
arr1
```

```python
arr1[1]
```

Stack a 1D array (1–5) into 5 rows (using tile or repeat).


```python
arr2 = np.arange(1,6)
np.tile(arr2, (5,1))
```

```python
np.repeat([arr2], 5,axis=0)
```

Convert a 5×5 matrix into upper-triangular form.


```python
arr3 = np.arange(1,26).reshape(5,5)
arr3
```

```python
np.triu(arr3)
```

Rotate a 3×3 matrix by 90 degrees using transpose and slicing.


```python
rotate = np.arange(1,10).reshape(3,3)
rotated = np.transpose(rotate)[::-1]
rotated
```

Concatenate three arrays [1,2], [3,4], [5,6] into one.


```python
a = np.array([1,2])
b = np.array([3,4])
c = np.array([5,6])
concated = np.concat((a,b,c))
concated
```

Horizontally stack an array [1,2,3] with itself three times.


```python
arr = np.array([1,2,3])
np.hstack([arr,arr,arr])
```

Split numbers 1–30 into 3 arrays of equal size.


```python
np.split(np.arange(1,31), indices_or_sections=3)
```
