```python
# Final dataset generator with imposed relationships

np.random.seed(42)

n_rows = 150000

# Locations (same as you provided) but we'll create popularity weights
locations = np.array([
    'Palam Vihar', 'Shastri Nagar', 'Khandsa', 'Central Secretariat',
    'Ghitorni Village', 'AIIMS', 'Vaishali', 'Mayur Vihar',
    'Noida Sector 62', 'Rohini', 'Udyog Bhawan', 'Vidhan Sabha',
    'Patel Chowk', 'Malviya Nagar', 'Madipur', 'Jama Masjid',
    'IGI Airport', 'Vinobapuri', 'Kashmere Gate', 'Pitampura',
    'Punjabi Bagh', 'Greater Noida', 'Tis Hazari', 'Noida Sector 18',
    'Kanhaiya Nagar', 'Okhla', 'Cyber Hub', 'Sadar Bazar Gurgaon',
    'Shastri Park', 'Faridabad Sector 15', 'Qutub Minar', 'Mundka',
    'DLF City Court', 'New Colony', 'Nirman Vihar',
    'New Delhi Railway Station', 'Civil Lines Gurgaon', 'Seelampur',
    'Noida Extension', 'Adarsh Nagar', 'Karol Bagh', 'Sultanpur',
    'Moti Nagar', 'Dilshad Garden', 'Aya Nagar', 'Rajiv Chowk',
    'MG Road', 'Jasola', 'Ardee City', 'Meerut', 'Anand Vihar ISBT',
    'Lajpat Nagar', 'Tughlakabad', 'Karkarduma', 'Dwarka Mor',
    'Anand Vihar', 'Uttam Nagar', 'Model Town', 'Mandi House',
    'Ghitorni', 'Rajiv Nagar', 'Sohna Road', 'Bahadurgarh',
    'Moolchand', 'INA Market', 'Sushant Lok', 'IIT Delhi',
    'Satguru Ram Singh Marg', 'Munirka', 'IGNOU Road', 'Akshardham',
    'Chhatarpur', 'IMT Manesar', 'Azadpur', 'Jhilmil', 'Sikanderpur',
    'Rohini West', 'Kashmere Gate ISBT', 'Kherki Daula Toll',
    'IFFCO Chowk', 'Hauz Rani', 'Sarai Kale Khan', 'Vishwavidyalaya',
    'Maidan Garhi', 'Yamuna Bank', 'Khan Market', 'Pragati Maidan',
    'Mansarovar Park', 'Rithala', 'Saidulajab', 'Nawada', 'Shahdara',
    'Tilak Nagar', 'DLF Phase 3', 'Laxmi Nagar', 'Huda City Centre',
    'Mehrauli', 'Botanical Garden', 'Ramesh Nagar', 'Dwarka Sector 21',
    'Inderlok', 'Rajouri Garden', 'Bhiwadi', 'Subhash Chowk',
    'Rohini East', 'Peeragarhi', 'Vasant Kunj', 'Pulbangash',
    'Kaushambi', 'Bhikaji Cama Place', 'Kirti Nagar', 'Paharganj',
    'Subhash Nagar', 'Old Gurgaon', 'Udyog Vihar', 'Hauz Khas',
    'Paschim Vihar', 'Ambience Mall', 'Lal Quila'
])

# Give higher weight to a subset of popular locations (simulate hotspots)
popularity_scores = np.ones(len(locations))
# Boost some indexes to simulate popular hubs
popular_idx = np.random.choice(len(locations), size=15, replace=False)
popularity_scores[popular_idx] = np.random.uniform(3, 8, size=popular_idx.size)
pop_weights = popularity_scores / popularity_scores.sum()

booking_statuses = ['Completed', 'No Driver Found', 'Incomplete', 'Cancelled by Driver', 'Cancelled by Customer']
# Completed majority
booking_status_weights = np.array([0.75, 0.08, 0.03, 0.07, 0.07])
vehicle_types = ['eBike', 'Go Sedan', 'Auto', 'Premier Sedan', 'Bike', 'Go Mini', 'Uber XL']
# define vehicle class multipliers (economy -> premium)
vehicle_multiplier = {
    'eBike': 0.6,
    'Bike': 0.6,
    'Auto': 0.8,
    'Go Mini': 0.9,
    'Go Sedan': 1.0,
    'Uber XL': 1.5,
    'Premier Sedan': 2.0
}
payment_methods = ['UPI', 'Debit Card', 'Cash', 'Uber Wallet', 'Credit Card']

# Create DataFrame scaffolding
df = pd.DataFrame(index=range(n_rows))

# Sample pickup and drop with popularity weights (ensure not equal)
pickup = np.random.choice(locations, size=n_rows, p=pop_weights)
drop = np.random.choice(locations, size=n_rows, p=pop_weights)
# Ensure pickup != drop
mask = pickup == drop
while mask.any():
    drop[mask] = np.random.choice(locations, size=mask.sum(), p=pop_weights)
    mask = pickup == drop
df['Pickup Location'] = pickup
df['Drop Location'] = drop

# Booking Status and Vehicle Type
df['Booking Status'] = np.random.choice(booking_statuses, size=n_rows, p=booking_status_weights)
df['Vehicle Type'] = np.random.choice(vehicle_types, size=n_rows, p=[0.12,0.25,0.08,0.08,0.12,0.20,0.15])  # example distribution

# --- Date and Time: impose monthly and hourly peaks ---
# Months: heavier booking in Mar-May and Sep-Nov (example seasonal peaks)
months = np.arange(1,13)
month_weights = np.array([0.06,0.07,0.10,0.10,0.10,0.08,0.06,0.08,0.10,0.10,0.08,0.07])
month_weights = month_weights / month_weights.sum()
chosen_months = np.random.choice(months, size=n_rows, p=month_weights)

# Day of month according to month (simple)
days = np.array([np.random.randint(1, 28+1) for _ in range(n_rows)])  # avoids month-end complexity
years = np.full(n_rows, 2024)
# Hours: peaks around 8-10 and 17-20
hour_probs = np.array([
    0.02, 0.01, 0.01, 0.01, 0.02, 0.03, 0.04, 0.06,
    0.08, 0.07, 0.03, 0.03, 0.04, 0.03, 0.03, 0.04,
    0.05, 0.07, 0.06, 0.05, 0.05, 0.04, 0.02, 0.02
])
hour_probs = hour_probs / hour_probs.sum()  # normalize

hours = np.random.choice(np.arange(0, 24), size=n_rows, p=hour_probs)

minutes = np.random.randint(0,60, size=n_rows)
seconds = np.random.randint(0,60, size=n_rows)
dates = pd.to_datetime({
    'year': years,
    'month': chosen_months,
    'day': days,
    'hour': hours,
    'minute': minutes,
    'second': seconds
})
df['Date'] = dates.dt.date.astype(str)
df['Time'] = dates.dt.time.astype(str)




# --------- Imposed relations for continuous variables ----------
# Ride Distance: base exponential, longer on weekends slightly, and depends on pickup-drop distinctiveness
base_dist = np.random.exponential(scale=6.0, size=n_rows)  # base distance in km
# bump distance if pickup-drop are less central (use location popularity as proxy)
loc_pop_inverse = 1.0 / (1.0 + popularity_scores)  # small for popular -> smaller distances
# Map each pickup location to its pop inverse
loc_to_popinv = dict(zip(locations, loc_pop_inverse))
pickup_popinv = np.vectorize(loc_to_popinv.get)(df['Pickup Location'])
drop_popinv = np.vectorize(loc_to_popinv.get)(df['Drop Location'])
suburb_factor = (pickup_popinv + drop_popinv) / 2.0
# Weekend multiplier
# Recreate weekend logic directly from Date column
is_weekend = pd.to_datetime(df['Date']).dt.weekday.isin([5,6]).astype(int)
df['Ride Distance'] = base_dist * (1 + suburb_factor) * (1 + 0.10 * is_weekend)

# ensure no zero distances
df['Ride Distance'] = df['Ride Distance'].clip(lower=0.2)

# Booking Value: base fare + distance * vehicle multiplier, plus peak-hour surge (hour-based)
base_fare = 25  # base booking platform currency units
vehicle_mult_array = np.array([vehicle_multiplier[v] for v in df['Vehicle Type']])
# surge: higher during peak hours
# Extract hour and month temporarily from Date & Time columns (not added to dataset)
hours = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
months = pd.to_datetime(df['Date']).dt.month

# Surge: higher during morning (8–10) and evening (17–20)
hour_surge = np.where(hours.isin([8,9,17,18,19,20]), 1.25, 1.0)

# Slight seasonal multiplier for busy months (Mar–May, Sep–Nov)
month_season = 1 + (months.isin([3,4,5,9,10,11]) * 0.05)

df['Booking Value'] = (base_fare + df['Ride Distance'] * 10 * vehicle_mult_array) * hour_surge * month_season
# add small noise
df['Booking Value'] = df['Booking Value'] * (1 + np.random.normal(0, 0.05, size=n_rows))
df['Booking Value'] = df['Booking Value'].clip(lower=5)

# Payment Method probability depends on booking value & vehicle type (higher price => card)
prob_card = 1 / (1 + np.exp(- (df['Booking Value'] - 150) / 40))  # sigmoid around 150
# build payment choice with probabilities
rand_uniform = np.random.rand(n_rows)
df['Payment Method'] = np.where(rand_uniform < prob_card, 
                                np.random.choice(['Debit Card', 'Credit Card', 'UPI'], size=n_rows, p=[0.45,0.35,0.20]),
                                np.random.choice(['Cash', 'Uber Wallet'], size=n_rows, p=[0.9,0.1])
                               )

# Driver Ratings: base by vehicle type and booking status, penalized by Avg CTAT (higher CTAT -> lower rating)
# First, create Avg CTAT: completed rides faster, others slower
base_ctat_completed = np.random.normal(3.0, 0.8, size=n_rows)  # minutes driver to complete/close? adjust as needed
base_ctat_notcompleted = np.random.normal(7.0, 2.0, size=n_rows)
df['Avg CTAT'] = np.where(df['Booking Status']=='Completed', base_ctat_completed, base_ctat_notcompleted)
# add small dependence on distance (longer rides slightly larger CTAT)
df['Avg CTAT'] += np.log1p(df['Ride Distance']) * 0.4
df['Avg CTAT'] = df['Avg CTAT'].clip(lower=0.2)

# Now driver rating base
vehicle_rating_base = {
    'eBike': 4.2,
    'Bike': 4.2,
    'Auto': 3.8,
    'Go Mini': 4.0,
    'Go Sedan': 4.3,
    'Uber XL': 4.4,
    'Premier Sedan': 4.6
}
vbase = np.array([vehicle_rating_base[v] for v in df['Vehicle Type']])
# penalize by Avg CTAT: each extra minute above 3 reduces rating by 0.08
df['Driver Ratings'] = vbase - np.maximum(0, (df['Avg CTAT'] - 3.0)) * 0.08
# also penalize non-completed bookings
df['Driver Ratings'] = np.where(df['Booking Status']=='Completed', df['Driver Ratings'], df['Driver Ratings'] - 0.6)
# add noise
df['Driver Ratings'] += np.random.normal(0, 0.18, size=n_rows)
df['Driver Ratings'] = df['Driver Ratings'].clip(1.0, 5.0).round(2)

# Avg VTAT: time until rider accepts driver or wait to start: completed faster
base_vtat_completed = np.random.normal(2.0, 0.7, size=n_rows)
base_vtat_notcompleted = np.random.normal(6.0, 1.5, size=n_rows)
df['Avg VTAT'] = np.where(df['Booking Status']=='Completed', base_vtat_completed, base_vtat_notcompleted)
# small dependence on distance and hour (peak hour slight increase)
df['Avg VTAT'] += np.log1p(df['Ride Distance']) * 0.15
hours = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
df['Avg VTAT'] += (hours.isin([8,9,17,18]) * 0.2)

df['Avg VTAT'] = df['Avg VTAT'].clip(lower=0.1)

# Customer Rating: depends on Driver Ratings, negatively on VTAT, positively on vehicle premium and booking status
vehicle_customer_bonus = {
    'eBike': 0.0,
    'Bike': 0.0,
    'Auto': -0.1,
    'Go Mini': 0.0,
    'Go Sedan': 0.1,
    'Uber XL': 0.15,
    'Premier Sedan': 0.3
}
vcust = np.array([vehicle_customer_bonus[v] for v in df['Vehicle Type']])
# base relation: customer's rating roughly driver rating minus penalty by VTAT (per minute 0.12)
df['Customer Rating'] = df['Driver Ratings'] + vcust - (df['Avg VTAT'] - 2.0) * 0.12
# completed bookings get small positive bump
df['Customer Rating'] = np.where(df['Booking Status']=='Completed', df['Customer Rating'] + 0.05, df['Customer Rating'] - 0.3)
# add small noise
df['Customer Rating'] += np.random.normal(0, 0.15, size=n_rows)
df['Customer Rating'] = df['Customer Rating'].clip(1.0, 5.0).round(2)

# Most booking time & month already imposed by sampling distributions
# Vehicle type relation with customer ratings implemented above
# Most pickup and drop location: compute top-k later if needed

# Introduce missing values randomly (5% for selected columns)
for col in ['Booking Value', 'Driver Ratings', 'Avg VTAT', 'Payment Method']:
    miss_idx = np.random.choice(df.index, size=int(0.05 * n_rows), replace=False)
    df.loc[miss_idx, col] = np.nan

# Introduce outliers in small fraction (1%) by multiplying some records
for col in ['Driver Ratings', 'Booking Value', 'Ride Distance', 'Avg VTAT']:
    idx = np.random.choice(df.index, size=int(0.01 * n_rows), replace=False)
    # For Driver Ratings keep within bounds but set some extreme low/high
    if col == 'Driver Ratings':
        df.loc[idx, col] = np.clip(df[col].mean() + np.random.normal(0, 2.5, size=len(idx)), 1, 5)
    else:
        df.loc[idx, col] = df[col].median() * np.random.uniform(5, 12, size=len(idx))

# Final cleanup: ensure types and sensible ranges
df['Booking Value'] = df['Booking Value'].round(2)
df['Ride Distance'] = df['Ride Distance'].round(2)
df['Avg VTAT'] = df['Avg VTAT'].round(2)
df['Avg CTAT'] = df['Avg CTAT'].round(2)
df['Driver Ratings'] = df['Driver Ratings'].round(2)
df['Customer Rating'] = df['Customer Rating'].round(2)

# For readability drop helper columns if you wish (Hour, Weekday kept for analysis)
# df = df.drop(columns=['Month','Hour','Weekday'])  # optional

# Quick checks: print basic distributions
print("Rows:", len(df))
print("Booking Status distribution:\n", df['Booking Status'].value_counts(normalize=True).head())
print("Top vehicle types:\n", df['Vehicle Type'].value_counts().head())
print("Payment method sample counts:\n", df['Payment Method'].value_counts().head())
print("Booking Value stats:\n", df['Booking Value'].describe())
print("Driver Ratings stats:\n", df['Driver Ratings'].describe())
print("Customer Ratings stats:\n", df['Customer Rating'].describe())

# Display top pickup / drop locations
print("Top pickup locations:\n", df['Pickup Location'].value_counts().head())
print("Top drop locations:\n", df['Drop Location'].value_counts().head())

# final dataframe
df.head()
```
