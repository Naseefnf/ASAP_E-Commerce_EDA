## Removing outliers

```python
q3 = np.percentile(df['Age'],75)
q1 = np.percentile(df['Age'],25)
q1,q3
```

```python
iqr = q3-q1
```

```python
lower = q1 - (1.5 * iqr)
upper = q3 + (1.5 * iqr)
lower, upper
```

```python
outliers = []
for i in df['Age']:
    if i<lower or i>upper:
        outliers.append(i)
outliers
```

```python
len(outliers)
```

```python
outlier_index = df[df['Age'].isin(outliers)].index
outlier_index
```

```python
df = df.drop(outlier_index, inplace=True)
```
