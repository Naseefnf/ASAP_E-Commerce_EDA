## Day 6: Hypothesis Testing

### 1. Null & Alternative Hypotheses

#### Null Hypothesis (H0)
        The default claimâ€”typically that there is no effect or no difference. We assume it is true until evidence suggests otherwise.

#### Alternative Hypothesis (H1)
        The claim youâ€™re testing forâ€”usually that there is an effect or a difference.

##### Example:

- H0: The average exam score this year equals last year's.

- H1: The average exam score this year is different from last year's.

### 2. p-value & Significance Level

#### p-value

        Probability of getting results as extreme as (or more extreme than) the observed data, assuming H0 is true.

#### Significance Level (Î±)

        The threshold for deciding if a result is "statistically significant"â€”often set at 0.05. If p-value < Î± reject H0.

##### Example:


```python
from scipy import stats
scores_2023 = [72, 68, 75, 70, 69]
scores_2024 = [74, 73, 77, 71, 72]
t_stat, p_value = stats.ttest_ind(scores_2023, scores_2024)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject H0: Significant difference")
else:
    print("Do not reject H0")
```

### 3. Type I and Type II Errors
- Type I Error: Rejecting H0 when it's actually true (a "false positive").

- Type II Error: Not rejecting H0 when it's actually false (a "false negative").
